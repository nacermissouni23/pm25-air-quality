- **Split `timestamp` into `utc` and `local` time features**

```python
import ast

# Parse the timestamp column (string representation of dict) and extract utc and local
df['timestamp'] = df['timestamp'].apply(ast.literal_eval)
df['utc'] = pd.to_datetime(df['timestamp'].apply(lambda x: x['utc']))
df['local'] = pd.to_datetime(df['timestamp'].apply(lambda x: x['local']))

# Drop the original timestamp column
df = df.drop('timestamp', axis=1)

# Display the result
df.head()
```

```python
df.info()
```

```python
sorted_df = df.sort_values(by='utc')
sorted_df.head()
```

```python
sorted_df.tail()
```

```python
# Extract just the date part (YYYY-MM-DD) from utc column
sorted_df['utc_date'] = pd.to_datetime(sorted_df['utc']).dt.date

# Get unique dates
unique_dates = sorted_df[['utc_date']].drop_duplicates()
unique_dates.columns = ['date']

# Save to CSV
unique_dates.to_csv('unique_dates.csv', index=False)

print(f"Found {len(unique_dates)} unique dates")
print(unique_dates.head())
```

```python
sorted_df['utc_date'] = sorted_df['utc'].dt.date.astype(str)
pm25_df = sorted_df.copy()

pm25_df.head()
```

- merge the datasets on `latitude`, `longitude`, `utc_date`

```python
gee['date'].value_counts()
```

```python
# Prepare sorted_df for merge: convert utc_date to string
sorted_df_merge = sorted_df.copy()
sorted_df_merge['date'] = sorted_df_merge['utc_date'].astype(str)

# Merge on date, latitude (lat), and longitude (lon)
merged_df = sorted_df_merge.merge(
    gee,
    on=['date', 'lat', 'lon'],
    how='inner'
)

print(f"Merged dataset shape: {merged_df.shape}")
print(f"PM2.5 rows: {len(sorted_df_merge)}, GEE rows: {len(gee)}, Merged rows: {len(merged_df)}")
print(f"\nColumns: {list(merged_df.columns)}")
print(f"\nFirst few rows:")
merged_df.head()
```

```python
import pandas as pd
import glob
import ast

LOCATIONS = [
    (36.1706, -5.3976, "Casablanca"), (36.7571, -4.5753, "Rabat"), (36.8422, -2.4612, "Tanger"),
    (37.16565, -3.60008, "Granada"), (37.60308, -0.97538, "Málaga"), (37.6936, -1.06464, "Motril"),
    (37.99359, -1.14472, "Vera"), (37.99102, -0.69008, "Mojacar"), (38.6394, -0.87238, "Almería"),
    (39.4776, -0.414, "Benidorm"), (39.7087, -0.2781, "Valencia"), (39.9986, -0.0731, "Castellón"),
    (39.769, 3.29, "Palma"), (40.00921, 3.8562, "Barcelona"), (39.26097, 9.13651, "Palermo"),
    (41.474, 2.0948, "Girona"), (41.11609, 1.19202, "Tarragona"), (43.41652, 5.22228, "Marseille"),
    (43.56248, 7.00722, "Antibes"), (42.67133, 9.43451, "Corsica"), (34.28899, 8.7616, "Tunis"),
    (35.28204, -2.94925, "Melilla"), (31.8747, -7.8907, "Laayoune"), (36.75649, 3.03772, "Algiers")
]

all_files = glob.glob("*.csv")
# Use the locations list to filter out unrelated CSV files like "here.csv" or "unique_dates.csv"
region_csvs = [f for f in all_files if any(region.replace(" ", "_") in f for lat, lon, region in LOCATIONS)]

final_data = []
print("--- Concatenating regional files ---")
for csv_file in region_csvs:
    try:
        df_region = pd.read_csv(csv_file)
        final_data.append(df_region)
        print(f"Loaded {len(df_region)} records from {csv_file}")
    except Exception as e:
        print(f"Error loading {csv_file}: {e}")

if final_data:
    combined_df = pd.concat(final_data, ignore_index=True)
    print(f"\nTotal hourly records combined: {len(combined_df)}")
    
    # Parse the string representation of dictionary to get UTC correctly
    print("Formatting timestamps and aggregating...")
    combined_df['timestamp'] = combined_df['timestamp'].apply(ast.literal_eval)
    combined_df['date'] = pd.to_datetime(combined_df['timestamp'].apply(lambda x: x['utc'])).dt.date
    
    daily_df = combined_df.groupby(['date', 'latitude', 'longitude']).agg({
        'pm25': 'mean',
        'sensor_name': 'first'
    }).reset_index()
    
    print(f"Total daily aggregated records: {len(daily_df)}")
    
    output_file = "pm25_1800days_final.csv"
    daily_df.to_csv(output_file, index=False)
    print(f"Saved aggregated data successfully to {output_file}")
else:
    print("No regional data found.")
```
